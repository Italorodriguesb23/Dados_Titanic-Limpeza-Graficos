```python
import pandas as pd
import seaborn as srn
import statistics as sts
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("train.csv")
```

```python
titanic.head()
```

```python
# Histograma da distribuição das idades
plt.hist(titanic['Idade'], bins=6, color='skyblue')
plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de barra de sobreviventes por gênero
sobreviventes_genero = titanic.groupby(['Genero', 'Sobreviveu']).size().unstack()

sobreviventes_genero.plot.bar(color=['red', 'skyblue'])

plt.title('Sobreviventes por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.show()
```

```python
# Gráfico de pizza de propoção de sobreviventes
proporcao_sobreviventes = titanic['Sobreviveu'].value_counts()
proporcao_sobreviventes.plot.pie(colors=['red', 'skyblue'], labels=['Não Sobreviveu', 'Sobreviveu'])

plt.title('Proporção de Sobreviventes')
plt.ylabel('')
plt.show()
```

```python
# Boxplot usando seaborn da distribuição de tarifas por classe
sns.boxplot(x='Classe', y='Tarifa', data=titanic, color='skyblue')
plt.title('Distribuição da Tarifa por Classe')
plt.xlabel('Classe')
plt.ylabel('Tarifa')
plt.show()
```

```python
# Gráfico de dispersão usando seaborn de tarifa vs idade
sns.scatterplot(x='Idade', y='Tarifa', data=titanic, color='skyblue')
plt.title('Dispersão: Tarifa vs Idade')
plt.xlabel('Idade')
plt.ylabel('Tarifa')
plt.show()
```

```python
#Gráfico de barras empilhadas de sobreviventes por gênero e classe
agrupado = titanic.groupby(['Genero', 'Classe', 'Sobreviveu']).size().unstack()
agrupado.plot(kind='bar', stacked=True, color=['red', 'skyble'])

plt.title('Sobreviventes por Gênero e Classe')
plt.xlabel('Gênero e Classe')
plt.ylabel('Quantidade')
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.tight_layout()
plt.show()
```
